## Analysis part begins from here

```python
data['Time taken to make changes'] = data["Revision date"]-data['First published']
data['Time taken to make changes'] = pd.to_numeric(data['Time taken to make changes'].dt.days, downcast='integer')
```

```python
print(data['Time taken to make changes'].max)
print(data[data['Time taken to make changes']!=0]['Time taken to make changes'].min)
```

```python
# graph 'Time taken to make changes' vs 'Index'
plt.scatter(data.index,data['Time taken to make changes'].fillna(0).astype(int))
```

```python
data['Stand alone time'] = data['First published']-data['Date added to the inventory']
data['Stand alone time'] = pd.to_numeric(data['Stand alone time'].dt.days, downcast='integer')
```

```python
print(data['Stand alone time'].max)
print(data[data['Stand alone time']!=0]['Stand alone time'].min)
```

```python
# graph 'Stand alone time' vs 'Index'
plt.scatter(data.index,data['Stand alone time'].fillna(0).astype(int))
```

```python
# alternate meds
diseases = data[['English common name of herbal substance','Use']].groupby('Use').nunique().reset_index()
diseases.columns = ['Use', 'Count']
diseases
```

```python
# graph of df diseases
plt.figure(figsize = (15,27))
ax = diseases.plot(x = 'Use', y= 'Count',kind = "bar" ,width =1)
```

```python
# diseases corresponding to new med added in the last 5 years with their no. of previous alternatives

med = data[['Use']].loc[data["First published"]>'2015-12-31'].reset_index().drop(columns = ["index"])
USE = diseases[['Use', 'Count']].reset_index().drop(columns = ['index'])
df = pd.merge(USE, med , how = 'inner')
df
```

```python
#graph 
fig1, ax1 = plt.subplots()
ax1.pie(df.Count, labels=df.Use, autopct='%1.1f%%',shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
